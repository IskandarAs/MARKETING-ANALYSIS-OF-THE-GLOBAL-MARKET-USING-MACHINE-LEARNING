>### Find the best way to sell a product

- In this section of the set problem we need to find one of the best way for sales product
- We are having 2 way to sell a product as Online & Offline, and we should find from that ways the best option, that have to be most comfortable and safety to business marketing

---
#### Cleaning & Preparing Data

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
sns.set_style('darkgrid')
df = pd.read_csv('Dataset.csv')
df.head()
```

```python
df['SHIP_DATE'] = pd.to_datetime(df['SHIP_DATE'])
df['year'] = df['SHIP_DATE'].dt.year
```

```python
total = df.groupby('REGION')[['UNITS_SOLD']].sum().reset_index()
total = total.sort_values(by='UNITS_SOLD', ascending=False)
total.style.background_gradient(cmap='YlGnBu', subset=['UNITS_SOLD'])
```

```python
year = df.groupby('year')[['UNITS_SOLD']].sum().reset_index()
year = year.sort_values(by='UNITS_SOLD', ascending=False)
year.style.background_gradient(subset = 'UNITS_SOLD', cmap ='RdBu')
```

---
#### Partition by sales channel

```python
f,ax1 = plt.subplots(figsize =(16,6.5))
sns.lineplot(x = "year", y = "UNITS_SOLD", hue = "SALES_CHANNEL", data = df,
            lw=3.2, palette="Set1",legend="brief")
plt.suptitle('Changining level of Units Sold of product year on year\n(2010 - 2017)', fontsize=14);
```

---
#### Partition by total profit

```python
f,ax1 = plt.subplots(figsize =(16,6.5))
sns.lineplot(x = "year", y = "TOTAL_PROFIT", hue = "SALES_CHANNEL", data = df,
            lw=3.2, palette="Set1",legend="brief")
plt.title("Changining level of Total Profit year on year\n(2010 - 2017)",fontsize=14);
```

---
#### Find the best option using different charts

```python
f, ax = plt.subplots(1,2, figsize=(16,6.5))
palette = [ "#E01E1B","#3791D7"]
sns.violinplot(x='year', y='UNITS_SOLD', hue='SALES_CHANNEL',palette=palette, data=df, ax=ax[0])

plt.suptitle('Changining level of Units Sold of product year on year\n(2010 - 2017)', fontsize=14)
sns.violinplot(x='year', y='UNITS_SOLD', hue='SALES_CHANNEL',palette='Set1', data=df, split=True, ax=ax[1])
plt.show()
```

```python
plt.subplots(figsize =(16,6.5))
sns.boxplot(x='year', y='UNITS_SOLD',hue='SALES_CHANNEL',data=df,palette='Set1')
plt.title("Changining level of Total Profit year on year\n(2010 - 2017)", fontsize=14);
```

```python
f, ax = plt.subplots(figsize=(16,6.5))
palette = [ "#E01E1B","#3791D7"]
plt.suptitle('Changining level of Units Sold of product year on year\n(2010 - 2017)', fontsize=14)
sns.barplot(x='year', y='UNITS_SOLD', hue='SALES_CHANNEL',palette =palette, data=df,estimator=lambda x: len(x) / len(df) * 100);
```

```python
f, ax = plt.subplots(1,2, figsize=(16,6.5))

colors = ["#D72626","#3791D7"]
labels ="Online", "Offline"

plt.suptitle('Changining level of Units Sold of product year on year\n(2010 - 2017)', fontsize=14)

df["SALES_CHANNEL"].value_counts().plot.pie(explode=[0,0.25], autopct='%1.2f%%', ax=ax[0], shadow=True, colors=colors, 
                                             labels=labels, fontsize=12, startangle=70)

ax[0].set_ylabel('Capacitance Units Sold of Product')
palette = [ "#E01E1B","#3791D7"]
sns.barplot(x="year", y="UNITS_SOLD", hue="SALES_CHANNEL", data=df, palette=palette, estimator=lambda x: len(x) / len(df) * 100)
ax[1].set(ylabel="Group by Year");
```

---
#### Partition by Region

```python
sns.set(rc={'figure.figsize':(25,25)}) 
g = sns.relplot(x='year', y='UNITS_SOLD', hue='SALES_CHANNEL', data=df,palette="Set1",
                col='REGION', col_wrap=2,
                height=4, aspect=1.6, style='SALES_CHANNEL',kind="line", sizes=(800,1000))
g.fig.suptitle('Changining level of Units Sold of product each Region year on year\n(2010 - 2017)' ,position=(0.5,1.05), fontweight='bold', size=15);
```

```python
plot = sns.catplot(x='year', y='UNITS_SOLD', hue='SALES_CHANNEL', data=df,
               kind='bar', col='REGION', col_wrap=3,height=4, palette="Set1",aspect=1.1)
plot.fig.suptitle('Changining level of Units Sold of product each Region year on year\n(2010 - 2017)' ,position=(0.5,1.05), fontweight='bold', size=15);
```
